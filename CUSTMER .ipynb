# 1.Over sampling
***In oversampling method  we use SMOTE(Synthetic minority oversampling Technique) is use.In this method we  work on minority class(increase the minority class).create new similar example for minority group***

```python
from imblearn.over_sampling import RandomOverSampler

# Separate features (X) and target (y)
X = df_copy1.drop('Will_Purchase', axis=1)
y = df_copy1['Will_Purchase']

# Initialize RandomUnderSampler
ros = RandomOverSampler(random_state=42)

# Fit and apply the resampler to the data
X_resampled, y_resampled = ros.fit_resample(X, y)

# Convert the resampled data back to a DataFrame
new_df = pd.concat([pd.DataFrame(X_resampled, columns=X.columns), pd.DataFrame(y_resampled, columns=['Will_Purchase'])], axis=1)


new_df.head()
```

# Kurtosis
- kurtosis is a statistical measure that describe the shape of a distribution to its peak.
- It provide insights into the "tailedness" or extremity of data points.
# Types of kurtosis
- Mesokurtic (kurtosis = 3) : The distribution has a normal shape.
- Leptokurtic (kurtosis > 3) : The distribution has heavy tails and a sharp peak.and it indicate more extreme outlier than a normal distribution.
- platykurtic (kurtosis < 3) : The distribution has light tails and a flatter peak. and fewer extreme outliers compared to a normal distribution.

# Skewness
- Skewness measures the asymmetry of a distribution. It can indicate whether the data is symmetric, or skewed to the left or right
# Types of skewness
- Skewness = 0: Perfectly symmetric distribution (e.g., normal distribution).
- Positive Skewness ( > 0): The tail on the right side is longer, and the bulk of data is concentrated on the left.
- Negative Skewness ( < 0): The tail on the left side is longer, and the bulk of data is concentrated on the right.

```python
from scipy.stats import skew,kurtosis
numerical_col=df.select_dtypes(include=['int64','float64']).columns
skewness=df[numerical_col].apply(skew).rename("skewness")
kurtosis_value=df[numerical_col].apply(kurtosis).rename("kurtosis")
distributionmatrix=pd.concat([skewness,kurtosis_value],axis=1)  
distributionmatrix
```

# ANOVA

# What is ANOVA?
- ANOVA stands for Analysis of Variance.
- ANOVA is a statistical test used to compare the means of two or more groups and determine if they are significantly different from each other.
- we can apply only categorical data.
- we can calculated by two methods:
- 1.F-statistics : The ratio of variance between group to variance within groups.
- 2.P_Value : Is the probability of obseving data no true differences between the group .
              if p_value < 0.05 i.e we reject the null hypothesis
              if p_value > 0.05 i.e  fail  to reject the null hypothesis.
# Types of ANOVA
1.ONE-WAY:
- There is only one independent variable and with more than two levels or groups.

2.TWO-WAY:
- Two-way ANOVA is two different categorical independent variables on one dependent variable.

```python
from scipy.stats import f_oneway # performs a one-way ANOVA test. for only one independent variable.
# define categorical column
categorical_col= ['Product_Interest']
for feature in categorical_col:
    groups = [df[df[feature] == category]['Annual_Spending'] for category in df[feature].unique()]
    f_stat, p_value = f_oneway(*groups) #unpacks the list of groups into separate arguments for the function.
    
    print(f"ANOVA results for {feature}:")
    print(f" F-statistic: {f_stat:.2f}")
    print(f" P-value: {p_value:.4f}\n")
```


**Product_Interest**
- F-statistic: 1.12
- P-value: 0.3376 . The p-value is  greater than 0.05 i.e The product interest has no significant effect. and we fail the reject the null hypothesis.

**Formulated Hypothesises:**
- Hypothesis 1: Higher loyalty score and online activity levels are associated with higher spending.
- Hypothesis 2: customer with shorter recency periods are more likely to purchase.
- Hypothesis 3: Income category has a significant influence on purchase likelihood.
- Hypothesis 4: product intersect is a key factor  influencing annual spending.

# T-test

- t-statistic is measure how strong the relationship is between the independent and dependent variable.

# Chi Square Test
- checking if two categorical variables are related or not.
- comparing observed vs expected frequencies.
- if there's a big diffrence between what's expected and what's observed that suggest the two variable might be.

```python
import pandas as pd
from scipy.stats import chi2_contingency  # contingency use for calculate the statistical significance of the relationship. 
# perform chi square test on two or more categorical variable
contingency_table=pd.crosstab(df['Income_Category'],df['Will_Purchase']) # crosstab for comparing two variable
chi2_test,p,Dof,expected=chi2_contingency(contingency_table)
print("p-value:",p) # check value is greater than 0.05 or not
if p<0.05:
    print("The variable are related")
else:
    print("The variable are not related")
print("chi square test:",chi2_test)
print("Degree of freedom:",Dof)  # Dof is number of independent values
print("Expected frequency:\n",expected)
```

- The Chi square Test is 1415.0081. i.e chi square test is large ,indicating strong association between Income category and will purchase.p-value is 1.1511e-309.

# Random Serach
- It randomly select values from the range of hyperparameter.
- Instend of trying every possible combination it sample , randomly from the define hyperparameter.

```python
# Random Search
from sklearn.model_selection import RandomizedSearchCV
import numpy as np

# Defining the parameter grid
param_dist = {
    'n_estimators': np.arange(100, 301, 50),
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [3, 5, 7],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Initializing RandomizedSearchCV
random_search = RandomizedSearchCV(
    estimator=GradientBoostingRegressor(random_state=42),
    param_distributions=param_dist,
    scoring='neg_mean_absolute_error',
    n_iter=20,  # Number of random combinations to test
    cv=3,
    verbose=2,
    n_jobs=-1,
    random_state=42
)

# Fit the model
random_search.fit(x_train, y_train)

# Best parameters and best score
best_params = random_search.best_params_
best_score = -random_search.best_score_  # Convert to positive
print("Best Parameters:", best_params)
print("Best MAE:", best_score)

```
